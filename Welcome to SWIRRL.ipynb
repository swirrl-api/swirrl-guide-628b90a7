# Welcome to SWIRRL!


SWIRRL is webservice API that allows Science Gateways to easily integrate analysis tools in their websites. By combining storage and tools, such as ***Workflows**, **Jupyter notebooks** and **GitHub** repositories, SWIRRL creates dedicated and reproducible working environments. 


## The /data folder

Thanks to the API's support for workflow and data-staging, SWIRRL workspaces are populated for you with data of interest,  collected from external data providers. If you are reading this page, it is likely that you have already requsted a web site to collect and provide you with a set of data in this notebook! 

SWIRRL manages all the data staged to the workspace in the ***/data*** folder. To make sure that your workspace is always consistent and your analysis reproducible, this is a 'read only' folder. However, you can always request to the web site to add more data for you here.

### /data/staginghistory

Each time data is staged to your workspace, SWIRRL keeps track of the changes by maintaining a staging history. You will find this in the ***/data/staginghistory*** folder.

<img src="images/staginghistory.png" alt="wfjob" style="height:260px;width:500px;"/>


Here, the ***stage.nnnn*** folders will preserve the state of your storage before the next staging workflow has been invoked. This will give you the opportuinity to always access previous version of the data. For instance, to reproduce analysis preformed on older versions fo the data and make comparisons.

***How to know whether a file has been updated?***

Files are available throughout the different stages, incrementally. Each ***stage.nnnn*** folder contains a ***swirrl_fileinfo.json***. Here you will find a complete record of metadata describing ***filename***, ***state***, ***id*** and ***sourceUrl*** of every of the file in your storage at a the given stage. The ***state** property can tell you whether the file is ***updated***, new, or unchanged.

## The SWIRRL Explorer

SWIRRL in combination with JupyterLab, allows you to customise your workspace, share and reproduce the progress of your analysis. You can update and install new python libraries in your environtment or restore it a previous state. You can freeze and create snapshot of your notebook directly to your Github account or to a repository managed by your project. 

You can control these actions via the SWIRRL Explorer. They will be logged and repurposed to you interactively. As shown in the the screenshot below, the list also inlcudes the workflows exectuted by the SWIRRL API, with a link to its detailed documentation.

### Update and Restore Libraries


SWIRRL and Jupyter allow the installation and update of additional python libraries in a workspace. This can be performed via the API or directly from the users via the Jupyter terminal. The image above shows that the ***Update Library*** activty also displays a summary of the installation performed, "*pip -install prov*". 

However, it may happen that changes to your environment could make it unstable or not suitable for your purposes. By clicking on the ***Restore*** button displayed close to a ***Create***, ***Update Libraries*** activities, users can recover from these situation by restoring the environment to a previous state. 


SWIRRL will always show the *diff* between the current environment and the one to be resotred. Upon confirmation, the curred workspace will replaced. The image below shows an example of the *diff* message shown to the user.


<img src="images/restore.png" alt="restore" style="height:270px;width:400px;"/>

**Note:** ***Restore*** activities are also logged, and SWIRRL allows you to recover to it.

### Snapshots

One  of  the  benefits  of  using  Jupyter  is  that  the  analysis methods can be documented in the notebook and shared  among  peers, who should be then able to execute them.  However, to  make  sure  that the  notebooks  are reproducible, the software environment and the source data needed for the analysis require to be recorded and managed. This is typically on the responsibility of the author of the notebook. 

SWIRRL instead automates this by offering a service that generates for you a reproducible snapshot of the current state of the workspace. 

A snapshot is a GitHub repository called ***Binder***. It includes your notebooks and files, the information about the software environment and the references to the data files in your ***/data*** folder. You can ask SWIRRL to create the repository directly in your account. Alternatively, it will use the default account associated with the website or the project hosting the system. Below we show how we create the snapshot *"tutorial-reanalysis"* via the SWIRRL Explorer.


<img src="images/snap.png" alt="snap" style="height:320px;width:400px;"/>


The advantage of creating ***Binders***, is that thes can be shared among peers who can redeploy them in http://mybinder.org. SWIRRL adds to the repository the references to the data and means of access.  MyBinder will use these informationn to automatically download the data, preparing a consistent and executable copy of the workspace.
